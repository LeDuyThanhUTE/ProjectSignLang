```python
#Đặt tên dựa vào chữ cái đầu
Gesture =  {'A_': 'A',
            'B_': 'B',
            'C_': 'C',
            'D_': 'D',
            'E_': 'E',
            'F_': 'F',
            'G_': 'G',
            'H_': 'H',
            'I_': 'I',
            'J_': 'J',
            'K_': 'K',
            'L_': 'L',
            'M_': 'M',
            'N_': 'N',
            'O_': 'O'}
```

```python
#Gán biến cho đầu ra
Gesture_map =  {'A': 0,
                'B': 1,
                'C': 2,
                'D': 3,
                'E': 4,
                'F': 5,
                'G': 6,
                'H': 7,
                'I': 8,
                'J': 9,
                'K': 10,
                'L': 11,
                'M': 12,
                'N': 13,
                'O': 14}
Gesture_names =  {0:'A',
                  1:'B',
                  2:'C',
                  3:'D',
                  4:'E',
                  5:'F',
                  6:'G',
                  7:'H',
                  8:'I',
                  9:'J',
                  10:'K',
                  11:'L',
                  12:'M',
                  13:'N',
                  14:'O'}
```

```python
# Gán các biến truy cập vào vào đường dẫn
path = '/content/drive/MyDrive/Colab Notebooks/Project/Data'
models_path = 'models/File_train.hdf5'
rgb = False
imgSize = 100
```

```python
# Hàm xử lí ảnh resize vè 224x224 và chuyển về array
def process_img(path):
    img = Image.open(path)
    img = img.resize((imgSize, imgSize))
    img = np.array(img)
    return img
```

```python
def process_data(X_data, y_data):
    X_data = np.array(X_data, dtype = 'float32')
    if rgb:
        pass
    else:
        X_data = np.stack((X_data,)*1, axis=-1)
    X_data /= 255
    y_data = np.array(y_data)
    y_data = to_categorical(y_data)
    return X_data, y_data
```

```python
# Hàm duyệt thư mục ảnh để train
def img_search(path):
    X_data = [] 
    y_data = []
    for directory, subdirectories, files in os.walk(path):
        for file in files:
            if not file.startswith('.'):
                imgpath = os.path.join(directory, file)
                Gesture_names = Gesture[file[0:2]]
                print(Gesture_names)
                print(Gesture_map[Gesture_names])
                y_data.append(Gesture_map[Gesture_names])
                X_data.append(process_img(imgpath))
            else:
                continue
    X_data, y_data = process_data(X_data, y_data)
    return X_data, y_data
```

```python
# Chia dữ liệu theo tỉ lệ train:test là 75:25
X_data, y_data = img_search(path)
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.25, random_state=12, stratify=y_data)
```

```python
models_path = '/content/drive/MyDrive/Colab Notebooks/Project/Model/saved_model.hdf5'
model_checkpoint = ModelCheckpoint(filepath=models_path, save_best_only=True)
early_stopping = EarlyStopping(monitor='val_acc',
                               min_delta=0,
                               patience=10,
                               verbose=1,
                               mode='auto',
                               restore_best_weights=True)

# Khởi tạo model VGG16
model1 = VGG16(weights='imagenet', include_top=False, input_shape=(imgSize, imgSize, 3))
optimizer1 = optimizers.Adam()
base_model = model1

# Thêm các lớp bên train
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu', name='fc1')(x)
x = Dense(128, activation='relu', name='fc2')(x)
x = Dense(128, activation='relu', name='fc2a')(x)
x = Dense(128, activation='relu', name='fc3')(x)
x = Dropout(0.5)(x)
x = Dense(64, activation='relu', name='fc4')(x)
# Có 15 lớp ký tự 
predictions = Dense(15, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=predictions)
model.summary()


# Đóng các lớp base_model, chỉ train lớp trên mình đã thêm vào
for layer in base_model.layers:
    layer.trainable = False
# Hàm train 50 lần 
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, batch_size=128, validation_data=(X_test, y_test), verbose=1,callbacks=[early_stopping, model_checkpoint])
```

```python
# Lưu model dã train ra file
model.save('train.h5')
```
